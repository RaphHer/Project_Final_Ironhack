A Monte Carlo simulation is a model used to predict the probability of different outcomes when the intervention of random variables is present. A Monte Carlo simulation is used to tackle a range of problems in many fields including investing, business, physics, and engineering. It is also referred to as a multiple probability simulation.

[Investopedia](https://www.investopedia.com/terms/m/montecarlosimulation.asp)

```python
import yfinance as yf #Yahoo Finance 
import pandas as pd
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')gs
```

```python
start = dt.datetime(2011, 1, 1)
end = dt.datetime(2022, 11, 29)
stock_data = yf.download('^GSPC', start, end) #GSPC is the ticker for the S&P 500
```

```python
stock_data
```

```python
returns = stock_data['Adj Close'].pct_change()
daily_vol = returns.std()
```

```python
daily_vol
```

```python
T = 252
count = 0
price_list = []
last_price = stock_data['Adj Close'][-1]

price = last_price * (1 + np.random.normal(0, daily_vol))
price_list.append(price)

for y in range(T):
    if count == 251:
        break
    price = price_list[count]* (1 + np.random.normal(0, daily_vol))
    price_list.append(price)
    count += 1

plt.plot(price_list)
plt.show()
```

```python
NUM_SIMULATIONS =  10000
df = pd.DataFrame()
last_price_list = []
for x in range(NUM_SIMULATIONS):
    count = 0
    price_list = []
    price = last_price * (1 + np.random.normal(0, daily_vol))
    price_list.append(price)
    
    for y in range(T):
        if count == 251:
            break
        price = price_list[count]* (1 + np.random.normal(0, daily_vol))
        price_list.append(price)
        count += 1
        
    df[x] = price_list
    last_price_list.append(price_list[-1])
        
fig = plt.figure()
fig.suptitle("Monte Carlo Simulation: S&P 500")
plt.plot(df)
plt.xlabel('Day')
plt.ylabel('Price')
plt.show()
```

```python
print("Expected price: ", round(np.mean(last_price_list),2))
print("Quantile (5%): ",np.percentile(last_price_list,5))
print("Quantile (95%): ",np.percentile(last_price_list,95))
```

```python
plt.hist(last_price_list,bins=100)
plt.title('Monte Carlo Distribution')
plt.axvline(np.percentile(last_price_list,5), color='r', linestyle='dashed', linewidth=2)
plt.axvline(np.percentile(last_price_list,95), color='r', linestyle='dashed', linewidth=2)
plt.show()
```
