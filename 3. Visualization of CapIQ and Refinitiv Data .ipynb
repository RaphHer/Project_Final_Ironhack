```python
import pandas as pd
import seaborn as sns; sns.set_theme(color_codes=True)
import matplotlib.pyplot as plt
```

```python
#Â Configuration to set so that all the Seaborn figs come out with this size
sns.set_context("poster")
sns.set(rc={"figure.figsize": (16.,8.)})
sns.set_style("whitegrid")
```

###  GDP US 1947-2022 Quarterly basis 

```python
df_gdp_us = pd.read_csv(r".\cleaned_data\01. CapIQ & Refinitiv\df_gdp_us.csv") # open the dataframe
df_gdp_us['Period'] = df_gdp_us["Period"].astype('datetime64[ns]') # change the date from object to datatime - important for visalization

f, ax = plt.subplots(figsize=(16, 8)) # Resetting the size of the figure
ax.set(xscale="log", yscale="log") # I added this to have the log scale 
sns.lineplot(data=df_gdp_us, x="Period", y="GDP-Quarterly ()").set(title='GDP US 1947-2022') #lineplot

plt.title('GDP US 1947-2022', fontsize=20)
plt.xlabel('Date', fontsize=16);
plt.ylabel('GDP', fontsize=16);
```

###  S&P 500 1928-2022 Daily basis 


```python
df_SP500 = pd.read_csv(r".\cleaned_data\01. CapIQ & Refinitiv\df_SP500.csv")
df_SP500['Period'] = df_SP500["Period"].astype('datetime64[ns]')
sns.lineplot(data=df_SP500, x="Period", y="S&P 500 index")

plt.title('S&P 500 1928-2022 Daily', fontsize=20)
plt.xlabel('Date', fontsize=16);
plt.ylabel('Index Value', fontsize=16);
```

### US Consumer Price Index 1914-2022 Monthly Basis 

```python
df_cpi_us = pd.read_csv(r".\cleaned_data\01. CapIQ & Refinitiv\df_cpi_us.csv")
df_cpi_us['Period'] = df_cpi_us["Period"].astype('datetime64[ns]')
sns.lineplot(data=df_cpi_us, x="Period", y="CPI (Y-o-Y change)")

plt.title('US Consumer Price Index 1914-2022 Monthly', fontsize=20)
plt.xlabel('Date', fontsize=16);
plt.ylabel('Value (%) Y-o-Y', fontsize=16);
```
